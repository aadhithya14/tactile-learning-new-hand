```python
import glob
import sys
import os
```

```python
# Notebook to stack videos


# First find all the video names given the task
task_name = 'plier_picking'
# Get the directories with the name plier_picking - traverse in eval_video and train_video
roots = [
    '/home/irmak/Workspace/tactile-learning/online_training_outs/eval_video/videos',
    '/home/irmak/Workspace/tactile-learning/online_training_outs/train_video/videos'
]
videos_roots = []

```

```python
for root in roots:
    video_root = glob.glob(os.path.join(root, f'*{task_name}*'))
    print('video_root: {}'.format(video_root))
    videos_roots += video_root
```

```python
len(videos_roots)
```

```python
# Get all the video paths
all_video_paths = []
for video_root in videos_roots:
    video_paths = glob.glob(os.path.join(video_root, '*.mp4'))
    all_video_paths += video_paths

```

```python
len(all_video_paths)
```

```python
def create_xstack_filter(num_vertical, num_horizontal, stack_order='horizontal'):
    # If stack_order is horizontal then the videos will first stack horizontally
    xstack_filter = ''
    if stack_order == 'horizontal':
        for v in range(num_vertical):
            vertical_addition = '0' if v == 0 else '+'.join([f'w{(i)*num_horizontal}' for i in range(v)])
            # print('vertical_addition: {}'.format(vertical_addition))
            for h in range(num_horizontal):
                horizontal_addition = '0' if h == 0 else '+'.join([f'h{(i)}' for i in range(h)])
                # print('horizontal_addition: {}'.format(horizontal_addition))

                curr_filter = f'{vertical_addition}_{horizontal_addition}|'
                xstack_filter += curr_filter

    elif stack_order == 'vertical':
        for h in range(num_horizontal):
            horizontal_addition = '0' if h == 0 else '+'.join([f'h{(i)*num_vertical}' for i in range(h)])
            # print('horizontal_addition: {}'.format(horizontal_addition))
            for v in range(num_vertical):
                vertical_addition = '0' if v == 0 else '+'.join([f'w{(i)}' for i in range(v)])
                # print('vertical_addition: {}'.format(vertical_addition))

                curr_filter = f'{vertical_addition}_{horizontal_addition}|'
                xstack_filter += curr_filter

    return xstack_filter[:-1] # Dont get the last |
```

```python
# Merge and dump the first 30/40 of the videos
first_dump = all_video_paths[:80]
xstack_filter = create_xstack_filter(10,8,'vertical')
input_strs = ' '.join([f'-i {video_path}' for video_path in first_dump])
vstack_strs = ''.join([f'[{i}:v]' for i in range(80)])
ffmpeg_str = f'ffmpeg {input_strs} -filter_complex "{vstack_strs}xstack=inputs=80:layout={xstack_filter}" output.mp4'
os.system(ffmpeg_str)

# ffmpeg -i input0 -i input1 -i input2 -i input3 -i input4 -i input5 -i input6 -i input7 -i input8 -i input9 -i input10 -i input11 -i input12 -i input13 -i input14 -i input15 -i input16 -filter_complex "[0:v][1:v][2:v][3:v][4:v][5:v][6:v][7:v][8:v][9:v][10:v][11:v][12:v][13:v][14:v][15:v]xstack=inputs=16:layout=0_0|w0_0|w0+w1_0|w0+w1+w2_0|0_h0|w4_h0|w4+w5_h0|w4+w5+w6_h0|0_h0+h4|w8_h0+h4|w8+w9_h0+h4|w8+w9+w10_h0+h4|0_h0+h4+h8|w12_h0+h4+h8|w12+w13_h0+h4+h8|w12+w13+w14_h0+h4+h8" output.mp4
```
