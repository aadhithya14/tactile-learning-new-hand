```python
# Using geometric series to make sure the sum is equal to 1
N = 50
a = 2/3 # Starting element - the biggest cost
# We're trying to find r such that with each element n is a * r^n and all the sums will be eqial to 1
# a * ((r**N - 1) / (r - 1)) = 1
r = 1/1.2

y = [a*(r**(N-n)) for n in range(N)]
x = [n for n in range(N)]
plt.plot(x,y)
```
